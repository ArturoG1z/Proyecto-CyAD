```python
# columnas donde todos los datos son nulos o 0
columnas_nulas = []

for columna in desempeno_DF.columns:
    if desempeno_DF[columna].replace(0, np.nan).isnull().all():
        columnas_nulas.append(columna)

print(f'Columnas donde todos los datos son nulos o 0:')
for columna in columnas_nulas:
  print('\t' + columna)
```

```python
# df_original.shape get the dimensions of the dataset rows * columns
# cantidad de celdas, datos nulos y porcentaje de datos nulos

def dimensiones_y_nulos(df, columnas_nulas = []):
  rows, columns = df.shape
  print(f'El dataset tiene {rows} filas y {columns} columnas')
  total_celdas = rows * columns
  print(f'Dimension del dataset: {total_celdas} celdas')
  total_datos_nulos = df.drop(columns=columnas_nulas).isnull().sum().sum()
  total_datos_nulos += rows * len(columnas_nulas)
  print(f'Cantidad de datos nulos: {total_datos_nulos}')
  porcentaje_datos_nulos = (total_datos_nulos / total_celdas) * 100
  print(f'Porcentaje de datos nulos: {porcentaje_datos_nulos:.2f}%')

dimensiones_y_nulos(desempeno_DF, columnas_nulas)
```

Esots datos son asociados a las columnas de actividades asignadas en los dos parciales y el porcentaje de aprovados en el curso

```python
#  analizar los datos nulos y el porcentaje de datos nulos por columna
def analisis_datos_nulos_ceros(df, tipo='nulos'):
  if tipo == 'nulos':
    datos_nulos = df.isnull().sum()
  elif tipo == 'ceros':
    datos_nulos = (df == 0).sum()
  else:
    return 'Tipo de datos no valido'
  porcentaje_datos_tipo = (datos_nulos / len(df)) * 100
  datos_por_tipo_DF = pd.DataFrame({
      'Datos ' + tipo: datos_nulos,
      'Porcentaje de datos ' + tipo: porcentaje_datos_tipo
  })
  return datos_por_tipo_DF
```

```python
# analisis de outliers

def analisis_outliers(df, columnas_numericas, type='z-scores'):
  outliers_por_columna = {}
  for columna in columnas_numericas:
    # plt.figure(figsize=(10, 5))
    # df[columna].iplot(kind='box', title=columna)
    # analisis estadistico
    if type == 'z-scores':
      z_scores = (df[columna] - df[columna].mean()) / df[columna].std()
      outliers_zscores = df[np.abs(z_scores) > 3][columna]
      print(f'Outliers en la columna {columna}: (longitud {len(outliers_zscores)}) z-scores')
      # print conversión a lista
      print(outliers_zscores.tolist())
      # marcar los outliers en el grafico de caja con rojo
      plt.figure(figsize=(10, 5))
      df[columna].iplot(kind='box', title=columna, boxpoints='outliers')
      outliers_por_columna[columna] = outliers_zscores
      
    else:
      q1 = df[columna].quantile(0.25)
      q3 = df[columna].quantile(0.75)
      iqr = q3 - q1
      lower_bound = q1 - 1.5 * iqr
      upper_bound = q3 + 1.5 * iqr
      outliers = df[(df[columna] < lower_bound) | (df[columna] > upper_bound)][columna]
      print(f'Outliers en la columna {columna}: (longitud {len(outliers)}) IQR')
      print(outliers.tolist())
      plt.figure(figsize=(10, 5))
      df[columna].iplot(kind='box', title=columna, boxpoints='outliers')
      outliers_por_columna[columna] = outliers

  return outliers_por_columna


columnas_numericas = metadata_numericos_DF['Columna']
outliers = analisis_outliers(desempeno_DF, columnas_numericas)
```

Generando un nuevo dataset con los datos limpios con outliers

```python
# se exporta un dataset aun sin eliminar los outliers
ruta = '../Data/dataset_limpio_con_outliers.xlsx'
desempeno_DF.to_excel(ruta, index=False)
```

Eliminando los outliers

```python
# eliminar outliers que se obtuvieron en el analisis_outliers

cleaned_DF = desempeno_DF.copy()

for columna, valores in outliers.items():
  cleaned_DF = cleaned_DF[~cleaned_DF[columna].isin(valores)]

# se exporta un dataset sin outliers
ruta = '../Data/dataset_limpio_sin_outliers.xlsx'
cleaned_DF.to_excel(ruta, index=False)

# mostrar diagrama de caja sin outliers

for columna in columnas_numericas:
  plt.figure(figsize=(10, 5))
  cleaned_DF[columna].iplot(kind='box', title=columna)
```
